# Plotting Astra files

```python
from astra import Astra, run_astra_with_distgen
from astra.plot import plot_fieldmaps, plot_stats, plot_stats_with_layout
import os
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams['figure.figsize'] = (16,4)
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
SCRATCH_DIR = 'tmp671qaf07'
H5_FILENAME = '1e9d2ea10ad6d0e6604171c8daf1cfc8.h5'
```

```python
astra = Astra.from_archive(os.path.join(SCRATCH_DIR, H5_FILENAME))
```

```python
astra.plot_fieldmaps()
```

```python
plot_stats(astra)
```

```python
kwargs = {'figsize':(10,5)}

plot_stats_with_layout(astra, ykeys=['sigma_x', 'sigma_y'], ykeys2=['sigma_z'], 
                           xkey='mean_z', xlim=(0, 1.3), 
                           nice=True, 
                           include_layout=False,
                           include_labels=True, 
                           include_particles=True, 
                           include_legend=True, **kwargs)
```

```python
astra.plot(y=['sigma_x', 'sigma_y'], y2=['sigma_z'], x='mean_z', xlim=(0,1.3), figsize=(10,5))
```
