# Astra evaluations

## ASTRA

```python
from astra.evaluate import evaluate_astra
from tempfile import TemporaryDirectory
from concurrent.futures import ProcessPoolExecutor as Executor
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
ASTRA_IN = 'astra.in'
PARAMETER = 'maxe(4)'
MIN, MAX, NUM = -12, -8, 10

SCRATCH = TemporaryDirectory()
```

```python
settings = {PARAMETER: MIN}
parameters = np.linspace(MIN, MAX, NUM)
```

```python
def evaluate(value):
    try:
        settings_copy = settings.copy()
        settings_copy[PARAMETER] = value
        output = evaluate_astra(settings_copy, astra_input_file=ASTRA_IN, archive_path=SCRATCH.name)
        output[PARAMETER] = value
        output['Exception'] = False
    except:
        output = {}
        output[PARAMETER] = value
        output['Exception'] = True
    return output
```

```python
SCRATCH.name
```

```python
executor =  Executor()
outputs = list(executor.map(evaluate, parameters))
```

```python
os.listdir(SCRATCH.name)
```

## Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(outputs)
df
```

```python
filtered_df=df[~df['Exception']]
```

```python
filtered_df
```

# Bokeh

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

output_notebook()
```

```python
source = ColumnDataSource(filtered_df)
source.column_names
```

```python
hover = HoverTool()
hover.tooltips=[(name, '@' + name) for name in source.column_names]

plot = figure()
plot.scatter(x=PARAMETER, y='end_sigma_x', source=source, marker='x', size=10, color='green', legend_label=r'sigma_x')
plot.scatter(x=PARAMETER, y='end_sigma_y', source=source, marker='x', size=10, color='red', legend_label='sigma_y')

plot.legend.location = 'top_left'
plot.legend.click_policy = 'hide'

plot.xaxis.axis_label = PARAMETER + ', MV/m'
plot.yaxis.axis_label = 'sigma, m'

plot.add_tools(hover)
plot.grid.grid_line_dash = [6, 4]

show(plot)
```
